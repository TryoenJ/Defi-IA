# Choix de l'avatar

Idées :
- Si un "avatar_id" existe déjà avec un "order_requests" plus petit alors il faut réutiliser ce même "avatar_id"
- Pour chaque avatar déjà créé, on cherche la date à laquelle il a effectué ses requests. Si la plus petite de ces dates est supérieure à notre nouvelle date choisie aléatoirement, alors on réutilise l'avatar pour la nouvelle request. Sinon on créé un nouvel avatar.

```python
def verif_avatar(random_avatar, random_date, tab_old_param):
    good_avatar = True
    for i in tab_old_param.index:
        if (random_avatar==tab_old_param['avatar_name'][i]):
            if (random_date>tab_old_param['date'][i]):
                good_avatar = False
        if good_avatar == False : break
    return good_avatar
```

# Verification de l'unicité de la requête

```python
def unicite(new_params, tab_old_params):
    unique = True
    for i in tab_old_params.index:
        if ((new_params['avatar_name']==tab_old_params['avatar_name'][i]) and
            (new_params['language']==tab_old_params['language'][i]) and 
            (new_params['city']==tab_old_params['city'][i]) and 
            (new_params['date']==tab_old_params['date'][i]) and
            (new_params['mobile']==tab_old_params['mobile'][i])):
            unique = False
        if unique == False : break
    return unique
```
